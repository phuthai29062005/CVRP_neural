### Bài toán CVRP

#### Định nghĩa
CVRP là bài toán tối ưu hóa để tìm tập hợp các tuyến đường phục vụ tất cả khách hàng, bắt đầu và kết thúc tại kho (depot), sao cho:
- Mỗi khách hàng được phục vụ **đúng một lần**
- Tổng nhu cầu trên mỗi tuyến **không vượt quá dung tích xe**
- Tối thiểu hóa **tổng chi phí vận tải**

#### Thành phần bài toán

**1. Dữ liệu đầu vào (Input Data)**
- **Kho (Depot)**: Điểm xuất phát và kết thúc cho tất cả tuyến đường (thường là điểm 1 trong định dạng TSPLIB95)
- **Khách hàng (Customers)**: Tập hợp n khách hàng, mỗi khách hàng i có:
  - Vị trí: tọa độ $(x_i, y_i)$
  - Nhu cầu: $d_i$ (lượng hàng cần giao)
- **Phương tiện (Vehicles)**: Đội xe với dung tích **C** (tối đa lượng hàng có thể chở)
- **Khoảng cách**: Ma trận khoảng cách $c_{ij}$ giữa các điểm i và j

**2. Ràng buộc (Constraints)**
- Mỗi khách hàng được phục vụ **đúng một lần** bởi **đúng một xe**
- Tổng nhu cầu trên mỗi tuyến ≤ dung tích xe: $$\sum_{i \in \text{route}} d_i \leq C$$
- Mỗi tuyến **bắt đầu và kết thúc tại depot**
- Không có xe nào bị "vô tập" (nếu xe rời depot phải quay lại)

#### Hàm mục tiêu (Objective Function)

Cuộc thi áp dụng quy ước cân bằng hai mục tiêu:

$$\text{Objective Function} = c \times NV + TD$$

Trong đó:
- **NV** (Number of Vehicles): Số lượng xe sử dụng
- **TD** (Total Distance): Tổng khoảng cách di chuyển (km)
- **c**: Hệ số trọng số = **1000** (được xác định empirically)

**Ý nghĩa**:
- Hệ số 1000 thể hiện chi phí cố định (mua xe, lương tài xế, bảo trì) cao hơn chi phí biến đổi (xăng dầu)
- Ưu tiên **giảm số xe hơn giảm khoảng cách**

#### Ví dụ minh họa

Giả sử có hai giải pháp:

| Tiêu chí | Giải pháp A | Giải pháp B |
|----------|-----------|-----------|
| Số xe (NV) | 5 | 4 |
| Tổng khoảng cách (TD) | 2000 km | 2500 km |
| Objective | 1000×5 + 2000 = **7000** | 1000×4 + 2500 = **6500** ✅ |

→ **Giải pháp B tốt hơn** vì tiết kiệm 1 xe, mặc dù khoảng cách xa hơn 500 km.

## Ý tưởng đầu tiên: Adaptive Route Generation với Permutation NP

### Tổng quan phương pháp

Phương pháp này sử dụng biểu diễn **hoán vị (permutation)** kết hợp với **cơ chế thích ứng động** để xác định điểm kết thúc tuyến đường, nhằm tối ưu hóa số lượng xe và quãng đường tổng cộng.

### Biểu diễn giải pháp

#### 1. Hoán vị khách hàng (NP Permutation)
- Một **hoán vị π** của n khách hàng: $π = [c_1, c_2, ..., c_n]$
- Các khách hàng được sắp xếp theo thứ tự trong hoán vị này
- Ví dụ: $π = [3, 1, 5, 2, 4]$ với 5 khách hàng

#### 2. Biến tỉ lệ động (Adaptive Route Break Ratio)
- Một biến **p (probability)** thay đổi động trong khoảng $[0, 1]$
- Biến này xác định **xác suất kết thúc tuyến đường** tại mỗi vị trí
- Ban đầu: $p = p_0$ (giá trị khởi tạo)
- Thay đổi động dựa trên số lượng khách hàng được thêm vào tuyến hiện tại

#### 3. Cơ chế tạo tuyến đường
Quá trình xây dựng giải pháp từ hoán vị:

**Bước 1**: Khởi tạo
- Bắt đầu từ depot
- Chọn khách hàng đầu tiên từ hoán vị
- Khởi tạo $p = p_0$ (ví dụ: $p_0 = 0.1$)

**Bước 2**: Thêm khách hàng vào tuyến hiện tại
- Xem xét khách hàng tiếp theo trong hoán vị
- Kiểm tra: $\sum_{i \in \text{current route}} d_i + d_{\text{next}} \leq C$ (không vượt capacity)
- Nếu **có thể thêm**:
  - Thêm khách hàng vào tuyến
  - Tăng tỉ lệ kết thúc: $p = p + \Delta p(n_{\text{added}})$
  - Trong đó: $\Delta p(n) = ???$ (công thức tăng tỉ lệ cần xác định)

**Bước 3**: Quyết định kết thúc tuyến
- Sinh số ngẫu nhiên $r \in [0, 1]$
- Nếu $r < p$: **Kết thúc tuyến hiện tại**, bắt đầu tuyến mới
- Nếu $r \geq p$: **Tiếp tục thêm khách hàng** vào tuyến hiện tại
- Nếu không thể thêm (vượt capacity): **Bắt buộc kết thúc tuyến**

**Bước 4**: Lặp lại cho đến khi hết khách hàng
- Reset $p = p_0$ cho tuyến mới
- Tiếp tục bước 2 và 3

### Hàm Fitness

Fitness được định nghĩa là:

$$\text{Fitness} = 1000 \times NV + TD$$

Trong đó:
- **NV**: Số lượng tuyến (xe) được tạo ra
- **TD**: Tổng quãng đường của tất cả các tuyến
  $$TD = \sum_{k=1}^{NV} \left( \sum_{i \in \text{route}_k} d(i, i+1) \right)$$
- Mục tiêu: **Minimize Fitness**

### Ý tưởng cơ chế thích ứng

#### Tại sao tỉ lệ phải tăng?

Khi một tuyến đã có nhiều khách hàng:
- Xác suất kết thúc tuyến nên **tăng lên** (tránh tuyến quá dài)
- Càng thêm nhiều khách hàng → Tuyến càng "đầy" → Xác suất kết thúc càng cao

#### Công thức tăng tỉ lệ (cần xác định)

Một số ý tưởng:
1. **Tuyến tính**: $\Delta p = 0.05 \times n_{\text{added}}$ (tỉ lệ với số khách hàng)
2. **Hàm capacity**: $\Delta p = 0.1 \times \frac{\text{load}_{\text{current}}}{C}$ (dựa trên mức độ "đầy")
3. **Lô-ga-rít**: $\Delta p = 0.1 \times \log(n_{\text{added}} + 1)$ (tăng chậm dần)
4. **Hỗn hợp**: $\Delta p = 0.08 \times \frac{\text{load}_{\text{current}}}{C} + 0.02 \times n_{\text{added}}$

### Các thuật toán lai ghép (Crossover Operators)

Để tạo quần thể từ hoán vị, sử dụng **3 thuật toán lai ghép** khác nhau:

#### 1. Order Crossover (OX)
- Chọn một đoạn từ cha 1, sao chép vào con
- Điền phần còn lại từ cha 2 (giữ thứ tự)
- **Ưu điểm**: Giữ nguyên thứ tự tương đối của các khách hàng

**Ví dụ**:
```
Cha 1: [1 2 3 4 5 6 7 8]
Cha 2: [3 7 5 1 6 8 2 4]
Đoạn lấy từ Cha 1: vị trí 3-5 → [3 4 5]
Con: [6 8 2 | 3 4 5 | 1 7] (điền từ Cha 2)
```

#### 2. Partially Mapped Crossover (PMX)
- Chọn hai điểm cắt, tạo "bản đồ" tương ứng
- Áp dụng bản đồ để giải quyết xung đột
- **Ưu điểm**: Tốt cho các bài toán định tuyến

**Ví dụ**:
```
Cha 1: [1 2 3 4 5 6 7 8]
Cha 2: [3 7 5 1 6 8 2 4]
Điểm cắt: 2-5
Bản đồ: 3→5, 4→1, 5→6, 6→8
Con được tạo bằng áp dụng bản đồ
```

#### 3. Cycle Crossover (CX)
- Tìm các "chu kỳ" trong hai cha
- Xen kẽ các chu kỳ để tạo con
- **Ưu điểm**: Bảo toàn vị trí tương đối của các gen

**Ví dụ**:
```
Cha 1: [1 2 3 4 5 6 7 8]
Cha 2: [3 7 5 1 6 8 2 4]
Chu kỳ 1: 1→3→5→6→8→4→1
Chu kỳ 2: 2→7→2
Con: [1 7 3 4 5 6 2 8] (xen kẽ chu kỳ)
```

### Chiến lược thử nghiệm

1. **Huấn luyện 3 mô hình riêng biệt**, mỗi mô hình sử dụng một thuật toán lai ghép khác nhau
2. **So sánh kết quả** trên tập dữ liệu test:
   - Fitness trung bình
   - Số lượng xe
   - Tổng quãng đường
   - Thời gian tính toán

3. **Chọn thuật toán tốt nhất** dựa trên metric kết hợp

### Ưu và nhược điểm

#### Ưu điểm ✅
- Biểu diễn **đơn giản và dễ hiểu** (hoán vị)
- **Thích ứng động**: Tỉ lệ kết thúc thay đổi theo tình huống
- **Linh hoạt**: Dễ dàng kết hợp với các toán tử tiến hóa khác
- **Đáp ứng ràng buộc**: Luôn đảm bảo không vượt capacity

#### Nhược điểm ❌
- **Công thức tỉ lệ chưa rõ**: Cần nhiều thử nghiệm
- **Không tối ưu cục bộ**: Chưa có cơ chế cải thiện trong tuyến
- **Phụ thuộc vào thứ tự**: Chất lượng phụ thuộc nhiều vào hoán vị ban đầu

## Tính Fitness với Multiple Random Route Partitions

### Ý tưởng chính

Vì **cơ chế random** trong việc xác định điểm kết thúc tuyến đường (xác suất p), một hoán vị giống nhau có thể tạo ra nhiều cách phân chia tuyến khác nhau. Để đánh giá chính xác chất lượng của một hoán vị, chúng ta thực hiện **multiple random trials** để tìm ra fitness tốt nhất.

### Phương pháp đánh giá Fitness

#### 1. Đánh giá một hoán vị đơn lẻ

Cho một hoán vị $π = [c_1, c_2, ..., c_n]$:

**Thuật toán**:
```
Nhập: Hoán vị π, số lần trial T (5-10 lần)
Đầu ra: Best fitness cho hoán vị này

fitness_list = []

For t = 1 to T:
    // Tạo cách phân chia tuyến random
    routes = GenerateRoutesWithRandomPartition(π)
    
    // Tính fitness cho cách phân chia này
    fitness_t = CalculateFitness(routes)
    
    fitness_list.append(fitness_t)

best_fitness_π = min(fitness_list)
return best_fitness_π
```

**Giải thích**:
- Chạy lại quá trình tạo tuyến từ hoán vị **T lần** (T ∈ [5, 10])
- Mỗi lần chạy sẽ có các quyết định random khác nhau (xác suất p)
- Thu thập tất cả fitness values
- **Chọn fitness tốt nhất** từ T lần thử

#### 2. Ví dụ minh họa

Giả sử hoán vị $π = [2, 5, 1, 3, 4]$ với 5 khách hàng:

| Lần thử | Tuyến 1 | Tuyến 2 | NV | TD | Fitness |
|--------|---------|---------|----|----|---------|
| 1 | [2, 5] | [1, 3, 4] | 2 | 150 | 2150 |
| 2 | [2, 5, 1] | [3, 4] | 2 | 145 | **2145** ✅ |
| 3 | [2] | [5, 1, 3, 4] | 2 | 160 | 2160 |
| 4 | [2, 5] | [1, 3, 4] | 2 | 150 | 2150 |
| 5 | [2, 5, 1, 3] | [4] | 2 | 152 | 2152 |

**Best fitness**: $min(2150, 2145, 2160, 2150, 2152) = \textbf{2145}$

#### 3. Áp dụng cho quần thể sau Crossover

Sau khi thực hiện **lai ghép (crossover)** để tạo ra các con, chúng ta cũng áp dụng phương pháp tương tự:

**Quy trình đầy đủ**:

**Bước 1**: Tạo quần thể ban đầu
- Khởi tạo N hoán vị ngẫu nhiên

**Bước 2**: Đánh giá fitness cho quần thể ban đầu
```
For each permutation π in population:
    best_fitness[π] = EvaluatePermutationWithTrials(π, T=5-10)
```

**Bước 3**: Chọn lọc (Selection)
- Chọn các hoán vị tốt nhất dựa trên best_fitness

**Bước 4**: Lai ghép (Crossover)
- Áp dụng một trong 3 toán tử lai ghép (OX, PMX, CX)
- Tạo ra quần thể con

**Bước 5**: Đánh giá quần thể con
```
For each child permutation π in offspring_population:
    best_fitness[π] = EvaluatePermutationWithTrials(π, T=5-10)
```

**Bước 6**: Hợp nhất và lặp lại
- Kết hợp cha và con
- Chọn lọc để giữ lại N tốt nhất
- Lặp lại từ bước 3

### 4. Chiến lược giá trị T (số lần trial)

#### Quyết định số lần trial

| Giai đoạn | T (trials) | Lý do |
|----------|-----------|-------|
| **Ban đầu (gen 1-10)** | 10 | Cần khám phá kỹ không gian giải pháp |
| **Giữa (gen 11-30)** | 7-8 | Cân bằng giữa khám phá và khai thác |
| **Cuối (gen 31+)** | 5 | Tập trung vào khai thác tốt hơn |

**Tùy chọn**: Có thể thiết lập T **động** dựa trên độ hội tụ của quần thể:
- Nếu quần thể **chưa hội tụ** → $T = 10$ (khám phá)
- Nếu quần thể **đang hội tụ** → $T = 7$ (cân bằng)
- Nếu quần thể **đã hội tụ** → $T = 5$ (khai thác)

### 5. Pseudo-code đầy đủ

```
Algorithm: GA with Multiple Random Route Partitions

Input: 
  - CVRP instance (depot, customers, capacity)
  - Population size: N
  - Number of generations: G
  - Number of trials: T ∈ [5, 10]
  - Crossover operators: OX, PMX, CX

Output:
  - Best solution found

1. Initialize: population = {π₁, π₂, ..., πₙ} (random permutations)

2. For each π in population:
     best_fitness[π] = EvaluateWithTrials(π, T)

3. For generation g = 1 to G:
   
   a. Selection:
      selected = SelectTopN(population, N/2)
   
   b. Crossover (apply 3 operators in parallel):
      offspring_OX = ApplyCrossover(selected, OX, count=N/3)
      offspring_PMX = ApplyCrossover(selected, PMX, count=N/3)
      offspring_CX = ApplyCrossover(selected, CX, count=N/3)
   
   c. Evaluate offspring:
      For each π in (offspring_OX ∪ offspring_PMX ∪ offspring_CX):
          best_fitness[π] = EvaluateWithTrials(π, T)
   
   d. Merge and select:
      all_solutions = population ∪ offspring_OX ∪ offspring_PMX ∪ offspring_CX
      population = SelectTopN(all_solutions, N)

4. Return: argmin(best_fitness[π]) for π in population
```

### 7. Tối ưu hóa hiệu suất

#### Sử dụng Parallel Processing
```python
# Pseudo-code
def EvaluatePermutationWithTrials(π, T):
    results = Parallel(
        [GenerateRoutesAndCalculateFitness(π) for _ in range(T)]
    )
    return min(results)
```

#### Caching kết quả
- Lưu trữ các hoán vị đã đánh giá
- Nếu hoán vị trùng → dùng kết quả đã lưu thay vì tính lại

#### Early Stopping
- Nếu T lần trial đầu tiên tìm được fitness quá tốt → có thể dừng sớm
- Ví dụ: Nếu fitness < threshold_tốt → dừng không cần T trials

### 8. So sánh 3 Crossover Operators với phương pháp này

| Toán tử | Ưu điểm | Nhược điểm |
|--------|---------|-----------|
| **OX** | Giữ thứ tự, dễ hiểu | Có thể bị local optima |
| **PMX** | Tốt cho VRP, đa dạng | Phức tạp hơn |
| **CX** | Bảo toàn vị trí, ổn định | Chậm hơn |

Với **multiple random trials**:
- OX → có thêm cơ hội tìm cách phân chia tốt hơn
- PMX → tận dụng tốt tính ngẫu nhiên
- CX → đảm bảo ổn định với nhiều trials

## Ý tưởng thứ 2: Intelligent Segment-based Crossover với RL-guided Vertex Selection

### Tổng quan phương pháp

Phương pháp này kết hợp **phân tích đoạn gen trùng** giữa cha mẹ với **Reinforcement Learning** để chọn đỉnh tối ưu khi xây dựng gen con, nhằm tạo ra các giải pháp chất lượng cao hơn thông qua lai ghép thông minh.

### Phần 1: Phân tích Đoạn Gen Trùng (Segment Matching)

#### 1.1 Khái niệm đoạn gen trùng

Một **đoạn gen trùng (common segment)** giữa hai cha mẹ là một chuỗi khách hàng liên tiếp xuất hiện **cùng vị trí** trong cả hai hoán vị.

**Ví dụ**:
```
Cha:  [1 2 3 4 5 6 7 8]
Mẹ:   [1 2 3 9 10 6 7 8]

Đoạn trùng:
- Đoạn 1: [1 2 3] ở vị trí 0-2 ✅
- Đoạn 2: [6 7 8] ở vị trí (5,6,7) và (5,6,7) ✅
- Đoạn giữa: khác nhau [4 5] vs [9 10]
```

#### 1.2 Thuật toán xác định đoạn gen trùng

```
Input: Cha (parent1), Mẹ (parent2)
Output: Danh sách các đoạn gen trùng

common_segments = []
current_segment = []
start_pos_cha = 0
start_pos_mẹ = 0

For position i from 0 to n-1:
    
    // Kiểm tra xem phần tử ở vị trí i có giống nhau không
    If parent1[i] == parent2[i]:
        If current_segment là rỗng:
            start_pos_cha = i
            start_pos_mẹ = i
        
        current_segment.append(parent1[i])
    
    Else:
        If current_segment không rỗng:
            common_segments.append({
                'segment': current_segment,
                'pos_cha': [start_pos_cha, i-1],
                'pos_mẹ': [start_pos_mẹ, i-1],
                'length': len(current_segment)
            })
            current_segment = []

// Xử lý đoạn cuối cùng nếu có
If current_segment không rỗng:
    common_segments.append(...)

Return common_segments
```

#### 1.3 Ví dụ chi tiết

```
Cha:  [2 5 1 3 4 7 6 8]
Mẹ:   [2 5 9 3 4 7 6 8]

Vị trí 0: 2 == 2 → thêm vào segment → [2]
Vị trí 1: 5 == 5 → thêm vào segment → [2, 5]
Vị trí 2: 1 ≠ 9 → kết thúc segment 1

Common segment 1: [2, 5] (vị trí 0-1 ở cả hai cha mẹ)

Vị trí 2: 1 ≠ 9 → bắt đầu segment mới
Vị trí 3: 3 == 3 → thêm vào segment → [3]
Vị trí 4: 4 == 4 → thêm vào segment → [3, 4]
Vị trí 5: 7 == 7 → thêm vào segment → [3, 4, 7]
Vị trí 6: 6 == 6 → thêm vào segment → [3, 4, 7, 6]
Vị trí 7: 8 == 8 → thêm vào segment → [3, 4, 7, 6, 8]

Common segment 2: [3, 4, 7, 6, 8] (vị trí 3-7)

Kết quả: 2 đoạn gen trùng
```

### Phần 2: Xây dựng Gen Con (Child Generation)

#### 2.1 Cơ chế chọn đoạn gen

**Bước 1**: Xác định đoạn gen trùng từ cha mẹ

**Bước 2**: Quyết định sử dụng đoạn nào
```
For each common_segment in common_segments:
    
    // Random 0/1 để quyết định sử dụng segment này hay không
    use_segment = RandomChoice([0, 1])
    
    If use_segment == 1:
        // Giữ nguyên đoạn này trong con
        child_genome[segment_pos] = segment
    Else:
        // Sẽ được điền bằng đỉnh được chọn từ cha hoặc mẹ
        child_genome[segment_pos] = None  // Chưa xác định
```

#### 2.2 Ví dụ minh họa

```
Cha:  [2 5 | 1 3 4 | 7 6 8]
Mẹ:   [2 5 | 9 3 4 7 6 | 8]

Common segments: [2,5] và [8]

Random choice:
- Segment [2,5]: USE (1)
- Segment [8]: DON'T USE (0)

Con ban đầu: [2 5 | ? ? ? | ? ? 8]

Phần cần điền: [1, 3, 4, 7, 6, 9]
```

### Phần 3: Chọn Đỉnh Thông Minh (RL-guided Vertex Selection)

#### 3.1 Hai chiến lược chọn đỉnh

Khi cần điền một đỉnh vào vị trí tiếp theo trong con, có **hai chiến lược**:

**Chiến lược 1: Random Selection (Khám phá)**
- Chọn ngẫu nhiên một đỉnh còn lại
- Nhanh, đa dạng, nhưng có thể không tối ưu

**Chiến lược 2: RL-guided Selection (Khai thác)**
- Sử dụng Reinforcement Learning để chọn đỉnh "tốt nhất"
- Dựa trên **giá trị Q(state, action)** - mức độ "tốt" của việc nối đỉnh này
- Chậm hơn, nhưng chất lượng cao hơn

#### 3.2 Tham số Alpha (Exploration vs Exploitation)

**Định nghĩa**:
$$\alpha \in [0, 1]$$

**Ý nghĩa**:
- Xác suất chọn **random** (khám phá): $\alpha$
- Xác suất chọn **RL-optimal** (khai thác): $1 - \alpha$

**Cơ chế**:
```
For each position to fill in child:
    
    r = RandomFloat([0, 1])
    
    If r < α:
        // Chọn random
        next_vertex = RandomSelect(remaining_vertices)
    Else:
        // Chọn RL-optimal
        Q_values = []
        For each remaining_vertex v:
            Q_values.append(Q(current_state, v))
        
        next_vertex = argmax(Q_values)
```

**Hướng dẫn chọn α**:
- $\alpha = 1.0$: 100% random (giống crossover thông thường)
- $\alpha = 0.7-0.8$: Cân bằng (khuyến khích sử dụng)
- $\alpha = 0.3-0.5$: Tập trung vào RL
- $\alpha = 0.0$: 100% RL-optimal (có thể hội tụ sớm)

### Phần 4: Phương pháp RL - Q-Learning

#### 4.1 Định nghĩa State và Action

**State**: 
```
State = {
  'current_position': vị trí hiện tại trong route,
  'current_load': tổng nhu cầu của route hiện tại,
  'last_vertex': đỉnh cuối cùng được thêm vào,
  'remaining_capacity': C - current_load
}
```

**Action**:
```
Action = vertex v ∈ {các đỉnh còn lại chưa được chọn}
```

**Reward**:
```
Reward = {
  'positive': Nếu chọn v là tốt (gần v, thêm vào được)
  'negative': Nếu chọn v là xấu (xa v, vượt capacity)
}
```

#### 4.2 Q-Function và Update

**Q-Value định nghĩa**:
$$Q(s, a) = \text{Giá trị ước tính của việc chọn action } a \text{ ở state } s$$

**Interpretation**:
- $Q(s, a)$ cao → việc thêm đỉnh $a$ vào route hiện tại là **tốt** (gần, không vượt capacity)
- $Q(s, a)$ thấp → việc thêm đỉnh $a$ là **xấu** (xa, có thể vượt capacity)

**Update Rule** (sau khi chọn vertex v và thêm vào):
$$Q(s, a) \leftarrow Q(s, a) + \eta \times [R(s, a) + \gamma \times \max_a Q(s', a) - Q(s, a)]$$

Trong đó:
- $\eta$: Learning rate (tốc độ học)
- $R(s, a)$: Reward ngay lập tức
- $\gamma$: Discount factor (tầm nhìn xa)
- $s'$: State tiếp theo

**Công thức cơ bản** (cần thử nghiệm):

```
Candidate 1: Distance-based reward
R(s, a) = -distance(last_vertex, a)
  → Ưu tiên các đỉnh gần
  
Candidate 2: Capacity-aware reward
R(s, a) = {
    -distance(last_vertex, a)           if load + demand[a] <= C
    -distance(last_vertex, a) - 100     if load + demand[a] > C (penalty)
}
  → Ưu tiên đỉnh gần nhưng đảm bảo capacity
  
Candidate 3: Hybrid reward
R(s, a) = -w1*distance(last_vertex, a) - w2*demand[a]/C
  → Cân bằng khoảng cách và nhu cầu
  
Candidate 4: Route-efficiency reward
R(s, a) = -distance(last_vertex, a) - lambda*route_fragmentation(a)
  → Giảm số lượng route bằng cách thêm nhiều đỉnh
```

### Phần 5: Phương pháp Beta - Route Partitioning

#### 5.1 Tham số Beta (Route Breaking Probability)

**Định nghĩa**:
$$\beta \in [0, 1]$$

**Ý nghĩa**:
- Xác suất **kết thúc route** hiện tại: $\beta$
- Xác suất **tiếp tục thêm** đỉnh vào route: $1 - \beta$

#### 5.2 Cơ chế sử dụng Beta

Khi xây dựng route từ đoạn gen:

```
For each position i in child_genome:
    
    If child_genome[i] needs to be filled:
        
        // Bước 1: Chọn đỉnh (dùng Alpha)
        vertex_v = SelectVertex(α, remaining_vertices)
        
        // Bước 2: Kiểm tra capacity
        If current_load + demand[v] <= C:
            current_load += demand[v]
            current_route.append(v)
            
            // Bước 3: Quyết định kết thúc route (dùng Beta)
            u = RandomFloat([0, 1])
            
            If u < β:
                // Kết thúc route, bắt đầu route mới
                routes.append(current_route)
                current_route = []
                current_load = 0
            Else:
                // Tiếp tục thêm đỉnh
                Continue to next position
        
        Else:
            // Vượt capacity, bắt buộc kết thúc route
            routes.append(current_route)
            current_route = []
            current_load = 0
            Backtrack to add v to new route
```

### Phần 6: Quy trình sinh con hoàn chỉnh

```
Algorithm: Intelligent Segment-based Crossover

Input:
  - Parent 1, Parent 2
  - α (exploration rate)
  - β (route break probability)
  - T (số lần để thử cách phân chia)

Output:
  - Child genome

1. FindCommonSegments(parent1, parent2)
   → common_segments = [seg1, seg2, ...]

2. Initialize child genome:
   For each segment in common_segments:
       if RandomChoice([0,1]) == 1:
           child[segment_pos] = segment
       else:
           child[segment_pos] = None

3. Fill remaining positions:
   remaining_vertices = AllVertices - used_vertices
   
   For each position i where child[i] == None:
       
       // Chọn đỉnh dùng Alpha
       vertex_v = SelectVertexWithAlpha(α, remaining_vertices)
       
       // Kiểm tra capacity
       If current_load + demand[v] <= C:
           current_route.append(v)
           remaining_vertices.remove(v)
           
           // Quyết định route dùng Beta
           if RandomFloat() < β:
               EndRoute()
       else:
           EndRoute()
           retry vertex v in new route

4. For T iterations (5-10 lần):
       routes_T = GenerateRoutesFromGenome(child, β)
       fitness_T = CalculateFitness(routes_T)
   
   best_fitness = min(fitness_T)
   
5. Return: (child, best_fitness)
```
